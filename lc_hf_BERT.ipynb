```python
import json
from transformers import pipeline

# This is https://huggingface.co/bert-base-multilingual-cased
unmasker = pipeline('fill-mask', model='mybert')

tests = [
  "Hvordan går [MASK]?",
  "Det er [MASK] i dag.",
  "Kroppstemperatur på 40 grader betyr [MASK].",
  "[MASK] er veldig fin.",
  "Jeg liker å [MASK].",
  "Hva kan du seie om [MASK]?"
]

for t in tests :
  print('\n'+t)
  res = unmasker(t)
  for a in res :
    print(a['token_str'],a['score'])
```
